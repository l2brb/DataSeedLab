# Import librerie


```python
import pandas as pd  # Libreria per la manipolazione dei dati
import openpyxl  # Libreria per lavorare con file Excel in formato .xlsx
```

# Import dataset Dealflow Coordinator da file XLSX


```python
file_path = "/home/l2brb/main/StartupWalkover/data_origin/Deal Flow Coordinator.xlsx"
sheet_name = "Dealflow centrale"


df = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=4)
#df = df.iloc[:, 1:]
df.head()
```

## Cleaning step 1

```python

cols_to_drop = list(df.columns[0:6]) + [df.columns[25], df.columns[26]]

df_cleaned = df.drop(cols_to_drop, axis=1)
df_cleaned.head()
```

## Export stepone Dealflow coordinator cleaned 

```python
export_path = "/home/l2brb/main/StartupWalkover/data_preprocessing/stepone/Dealflow_coordinator_cleaned.csv"
df_cleaned.to_csv(export_path, index=False)
```

## Import Dataset DF_Scaleup

```python
file_path = "/home/l2brb/main/StartupWalkover/data_origin/DF_Scaleup.xlsx"
sheet_name = "Deal Flow"


df_scaleup = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=1)
df_scaleup = df_scaleup.iloc[:, 3:]
df_scaleup.head()
```

```python
print(df_scaleup.info())
```

```python
df_scaleup.shape
```

```python
df_scaleup.describe(include=['object'])
```

```python
df_scaleup = df_scaleup.dropna(subset=['ID'])

df_scaleup['ID'] = df_scaleup['ID'].fillna(0).astype(int)

print(df_scaleup.info())
```

```python
df_scaleup.shape
```

## Export stepone Dealflow scaleup cleaned 

```python
export_path = "/home/l2brb/main/StartupWalkover/data_preprocessing/stepone/Dealflow_scaleup_cleaned.csv"
df_scaleup.to_csv(export_path, index=False)
```

## Import DF_VV6-VV7

```python
file_path = "/home/l2brb/main/StartupWalkover/data_origin/DF_VV6-VV7.xlsx"
sheet_name = "Deal Flow"


df_VV6VV7 = pd.read_excel(file_path, sheet_name=sheet_name, skiprows=1)
df_VV6VV7 = df_VV6VV7.iloc[:, 3:]
df_VV6VV7.head()
```

```python
print(df_VV6VV7.info())
```

```python
df_VV6VV7 = df_VV6VV7.dropna(subset=['ID'])

df_VV6VV7['ID'] = df_VV6VV7['ID'].fillna(0).astype(int)

print(df_VV6VV7.info())
```

```python
df_VV6VV7.shape
```

## Export stepone Dealflow VV6 VV7 cleaned

## Merge DF_Scaleup e DF_VV6-VV7

```python
df_merged = pd.concat([df_VV6VV7, df_scaleup], axis=0, ignore_index=True)
print(df_merged.shape)
df_merged.head()

export_path = "/home/l2brb/main/StartupWalkover/data_preprocessing/stepone/Dealflow_merged_scaleup_vv6_vv7.csv"
df_merged.to_csv(export_path, index=False)
```

## Merge DF Coordinator & DF merged (Scaleup + vv6 and vv7)

```python
df_full = pd.merge(df_cleaned, df_merged, on="ID", how="inner")
print(df_full.head())
```
