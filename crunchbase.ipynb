# Import df_crunchbase

```python
file_path_1 = "/home/l2brb/main/StartupWalkover/data_preprocessing/export_crunchbase/first-100011-06-2025.csv"
file_path_2 = "/home/l2brb/main/StartupWalkover/data_preprocessing/export_crunchbase/1000-200011-06-2025.csv"
file_path_3 = "/home/l2brb/main/StartupWalkover/data_preprocessing/export_crunchbase/2000-300011-06-2025.csv"
file_path_4 = "/home/l2brb/main/StartupWalkover/data_preprocessing/export_crunchbase/last-11-06-2025.csv"

df_1 = pd.read_csv(file_path_1)
df_2 = pd.read_csv(file_path_2)
df_3 = pd.read_csv(file_path_3)
df_4 = pd.read_csv(file_path_4)

```

```python
df_1.shape
```

```python
df_1.head()
```

```python
#Merging di crunchbase
# Unisci i 4 dataframe verticalmente usando pd.concat
df_crunchbase = pd.concat([df_1, df_2, df_3, df_4], ignore_index=True)
```

```python
df_crunchbase.shape
```

```python
export_path = "/home/l2brb/main/StartupWalkover/data_preprocessing/export_crunchbase/full.csv"
df_crunchbase.to_csv(export_path, index=False)
```

# Merge con df Crunchbase

```python
merged = pd.merge(
    df,
    df_crunchbase,
    left_on="Progetto/Società",      # chiave nel 1° df
    right_on="Organization Name",    # chiave nel 2° df
    how="inner",                     # solo le corrispondenze esatte
    suffixes=("_it", "_cb"),
    indicator=True                   # facoltativo: ti mostra chi ha fatto match
)


```
